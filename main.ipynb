```python
import yfinance as yf
```

```python
AAPL = yf.download(tickers='AAPL',period='5y',interval = '1d')#take data apple company
```

```python
AAPL.drop("Adj Close", axis = 1, inplace =True)#delete column adj close
AAPL
```

```python
TSLA = yf.download(tickers='TSLA',period='5y',interval = '1d')#take data Tesla company
```

```python
TSLA.drop("Adj Close", axis = 1, inplace =True)#delete column adj close
TSLA
```

```python
GG = yf.download(tickers='GOOGL',period='5y',interval = '1d')#take data GOOGLE company
```

```python
GG.drop("Adj Close", axis = 1, inplace =True)#delete column adj close
GG
```

```python
list_company = ['AAPL','TSLA','GG']
```

```python
company_stocks = pd.concat([AAPL,TSLA,GG],axis = 1, keys =list_company)
company_stocks
```

```python
company_stocks.columns.names = ['Company Name','Stock Info']
company_stocks
```

```python
#max value close for every company
company_stocks.xs(key='Close',axis = 1,level='Stock Info').max()
```

```python
#min value close for every company
company_stocks.xs(key='Close',axis = 1,level='Stock Info').min()
```

```python
value_cp = pd.DataFrame()#ti xuat loi nhuan
value_cp.head()
```

```python
for name in list_company:
    value_cp[name+' close_value'] = company_stocks[name]['Close'].pct_change()*100

value_cp.columns.names = ['info_cp',]
value_cp.head()
```

```python
import seaborn as sns

sns.pairplot(value_cp);
```

```python
company_stocks1 = pd.concat([AAPL,TSLA,GG],axis =0, keys =list_company)
data_frame1 = pd.DataFrame()
data_frame1 = company_stocks1['Close']
data_frame1 = data_frame1.reset_index()
data_frame1.columns = ['Name','Date','Money']
print(data_frame1)
```

```python
g = sns.pairplot(data_frame1,hue = 'Name',x_vars='Date',y_vars = 'Money');
for ax in g.axes.flat[:2]:
    ax.tick_params(axis='x', labelrotation=45)
g
```

```python
for tick in list_company:
    company_stocks[tick]['Close'].plot(figsize=(12,4),label = tick)
```

```python
#Simple Moving Average (SMA): là đường trung bình động đơn giản được tính bằng trung bình cộng các mức giá đóng cửa trong một thời gian giao dịch nhất định
AAPL['Close'].loc['2020-01-01':'2021-01-01'].ta_plot(study = 'sma',periods=[7,14,21],title = 'SMA APPL')
```
